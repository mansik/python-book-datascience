# 그래프의 기본

* matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline  # 그래프를 주피터 노트북 내부에 출력(없어도 됨)
```

```python
x = [1, 2, 3, 4]
y = [2, 3, 4, 5]
```

```python
plt.plot(x, y)
plt.show()
```
