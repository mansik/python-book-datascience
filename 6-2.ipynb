# 그래프 꾸미기

* 그래프를 구성하는 객체: figure(axes를 감싼다.) > axes(subplot, subplots)(그래프를 감싼다) > 그래프(plot, pie,...)
* figure, axes는 생략가능

## 한글 설정

```python
%matplotlib inline
import matplotlib.pyplot as plt
import sys
from matplotlib import font_manager, rc # 폰트를 설정하기 위해서 사용
```

```python
if sys.platform == 'darwin': # Mac OS 이름 확인
    path = '/Library/Fonts/AppleGothic.ttf'
elif sys.platform == 'win32': # windows 이름 확인
    path = 'c:/Windows/Fonts/malgun.ttf'
else:
    print('Unknown system...')

font_name = font_manager.FontProperties(fname=path).get_name()
rc('font', family=font_name) # 한글 폰트 지정
plt.rcParams['axes.unicode_minus'] = False # 폰트를 한글로 설정하면 `-` 값이 처리 되지 않는 문제 해결
```

## 그래프를 구성하는 객체

* 그래프를 구성하는 객체 => figure(axes를 감싼다.) > axes(그래프를 감싼다) > 그래프

```python
# figure 객체 생성
fig = plt.figure(figsize=(8, 4))
```

## axes 추가하기

```python
# 그래프가 들어갈 axes 객체를 2개 생성
fig = plt.figure(figsize=(8, 4))  # 같은 cell에 존재해야 한다.
ax1 = fig.add_subplot(2, 1, 1) # 2행 1열의 1번째 axes, fig.add_subplot(211)도 가능
ax2 = fig.add_subplot(2, 1, 2) # 2행 1열의 2번째 axes
plt.show()  # 그래프 보기
```

```python
# 하나의 행에 두 개의 열을 그래프로 그리기
fig = plt.figure(figsize=(8,4))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
plt.show()
```

## 한번에 figure, axes 만들기

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
fig.show()
```

## 그래프 그리기

* 그래프의 형상 셋팅(color, marker, line style)
* Figure 객체 재사용(plt.gcf())

```python
f1 = plt.figure(figsize=(8,4))
plt.title(" Figure ")
plt.plot([1, 2, 3, 4], 'ro:') # 그래프 그리기, color=red, marker=o(circle), line style=:(spot) # plot 도움말 참고

f2 = plt.gcf() # get current figure
plt.plot([4, 6, 8, 10], 'bo:') # 그래프 그리기, color=blue
plt.show()
```

```python
# subplots로 figure와 axes 객체를 한번에 만들기
fig, axes = plt.subplots(figsize=(8,4)) # default 값인 nrows=1, ncols=1로 생략함
plt.title(" Figure Axes ")
axes.plot([1, 2, 3, 4], 'ro-') # line style=-(solid line style) # plot 도움말 참고

f2 = plt.gcf() # get current figure
axes.plot([4, 6, 8, 10], 'bo--')
plt.show()
```

## 제목을 한글로 표시

```python
import numpy as np

x = np.arange(1, 10)
y = x**2
```

```python
x, y
```

```python
fig = plt.figure(figsize=(8, 4))
plt.title("한글 제목")
plt.plot(x, y)
plt.show()
```

## 그래프 형상(color, marker, line style)

```python
fig = plt.figure(figsize=(8, 4))
plt.plot(x, y, 'ms-.')
plt.title("line dot 스타일의 plot")
plt.show()
```

## 스타일 매개변수 사용

```python
fig = plt.figure(figsize=(8, 4))
plt.title("라인 스타일의 plot")
plt.plot(x, y, color='r', marker='s', linestyle='--')
plt.show()
```

```python
fig = plt.figure(figsize=(8,4))
plt.title("스타일 적용 예제")
plt.plot(x, y, c="r", lw=3, ls="--", marker="s", ms=6, mec="b", mew=3, mfc="r") # 스타일 약어 사용, 도움말 참고
plt.show()
```

## 그래프 범위 지정(plt.xlim, plt.ylim = xlimit, ylimit)

* 축의 정보는 plot 함수에 전달하는 변수 x, y의 값을 이용해서 임의로 표시한다.
* 명시적으로 축의 범위를 지정한다.

```python
fig = plt.figure(figsize=(8,4))
plt.plot(x, y, c='r', lw=3, ls='--', marker='s', ms=6, mec='b', mew=3, mfc='r')
plt.title("x축, y축의 범위 설정")
plt.xlim(0, 10) # xlimit
plt.ylim(0, 100)
plt.show()
```

## 그래프 좌표에 대한 레이블 지정(plt.xticks, plt.yticks)

* 축에 표시되는 숫자를 명시적으로 표시

```python
fig = plt.figure(figsize=(8,4))
plt.plot(x, y, c='r', lw=3, ls='--', marker='s', ms=6, mec='b', mew=3, mfc='r')
plt.title("x축, y축의 tick 설정")
plt.xticks(x)
plt.yticks(y)
plt.show()
```

## 좌표에 그리드 지정(plt.grid(True))

```python
fig = plt.figure(figsize=(8,4))
plt.plot(x, y, c='r', lw=3, ls='--', marker='s', ms=6, mec='b', mew=3, mfc='r')
plt.title("x축, y축의 tick 설정")
plt.xticks(x)
plt.yticks(y)
plt.grid(True)
plt.show()
```

## 축의 제목 표시(plt.xlabel)

```python
fig = plt.figure(figsize=(8,4))
plt.plot(x, y, c='r', lw=3, ls='--', marker='s', ms=6, mec='b', mew=3, mfc='r')
plt.title("x축, y축의 tick 설정")
plt.xticks(x)
plt.yticks(y)
plt.grid(True)
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```
